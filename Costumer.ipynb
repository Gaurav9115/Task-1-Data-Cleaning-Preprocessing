# TASK 1: DATA CLEANING & PREPROCESSING

```python
import pandas as pd
```

### STEP 1: LOAD DATASET

```python
df = pd.read_csv(
    r"customers-100.csv"
)
```

```python
print("Original Shape:", df.shape)
```

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
print(df.columns.tolist())
```

### STEP 2: REMOVE DUPLICATES

```python
duplicate_count = df.duplicated().sum()
print("\nDuplicate Records:", duplicate_count)
```

```python
df = df.drop_duplicates()
```

```python
# Check duplicate customer IDs
customer_duplicates = df["Customer Id"].duplicated().sum()
print("Duplicate Customer IDs:", customer_duplicates)
```

### STEP 3: COLUMN MANAGEMENT

```python
# 'index' column is not needed
df.drop(columns=["Index"], inplace=True)
```

### STEP 4: CHECK MISSING VALUES

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Fill text columns with mode
text_columns = [
    "First Name",
    "Last Name",
    "Company",
    "City",
    "Country",
    "Phone 1",
    "Phone 2",
    "Email",
    "Website"
]
for col in text_columns:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Drop rows where customer_id is missing
df.dropna(subset=["Customer Id"], inplace=True)
```

### STEP 5: DATA TYPE CORRECTION

```python
df["Customer Id"] = pd.to_numeric(
    df["Customer Id"],
    errors="coerce"
)

df["Subscription Date"] = pd.to_datetime(
    df["Subscription Date"],
    errors="coerce"
)
```

### STEP 6: FORMAT STANDARDIZATION

```python
# Names

df["First Name"] = (
    df["First Name"]
    .astype(str)
    .str.strip()
    .str.title()
)

df["Last Name"] = (
    df["Last Name"]
    .astype(str)
    .str.strip()
    .str.title()
)

# Company

df["Company"] = (
    df["Company"]
    .astype(str)
    .str.strip()
)

# City

df["City"] = (
    df["City"]
    .astype(str)
    .str.strip()
    .str.title()
)

# Country

df["Country"] = (
    df["Country"]
    .astype(str)
    .str.strip()
    .str.title()
)

# Email

df["Email"] = (
    df["Email"]
    .astype(str)
    .str.strip()
    .str.lower()
)

# Website

df["Website"] = (
    df["Website"]
    .astype(str)
    .str.strip()
    .str.lower()
)

# Phone Numbers

df["phone 1"] = (
    df["Phone 1"]
    .astype(str)
    .str.replace(r"[^0-9]", "", regex=True)
)

df["Phone 2"] = (
    df["Phone 2"]
    .astype(str)
    .str.replace(r"[^0-9]", "", regex=True)
)
```

### STEP 7: VALIDATION CHECKS

```python
print("\nRemaining Missing Values:")
print(df.isnull().sum())

print("\nRemaining Duplicates:")
print(df.duplicated().sum())

assert df.duplicated().sum() == 0
```

### STEP 8: TEXT CLEANING

```python
text_cols = df.select_dtypes(include="object").columns
for col in text_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
    )
```

### STEP 9: SAVE CLEANED DATA

```python
df.to_csv(
    "cleaned_customers.csv",
    index=False
)

print("\nData Cleaning Completed Successfully")
print("Cleaned File Saved: cleaned_customers.csv")

print("\nFinal Shape:", df.shape)

print("\nPreview:")
print(df.head())
```
