# TASK 3: DATA CLEANING & PREPROCESSING

```python
import pandas as pd
```

### STEP: 1 LOAD DATASET

```python
df = pd.read_csv(
    r"people-100.csv"
)
```

```python
print("Original Shape:", df.shape)
```

```python
print(df.columns.tolist())
```

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

### STEP 2: REMOVE DUPLICATES

```python
duplicate_count = df.duplicated().sum()
print("\nDuplicate Records:", duplicate_count)
```

```python
df = df.drop_duplicates()
```

### STEP 3: COLUMN MANAGEMENT

```python
df = df.rename(columns={
    "Index": "record_id",
    "User Id": "user_id",
    "First Name": "first_name",
    "Last Name": "last_name",
    "Sex": "gender",
    "Email": "email",
    "Phone": "phone",
    "Date of birth": "date_of_birth",
    "Job Title": "job_title"
})

# Reorder columns
df = df[
    [
        "record_id",
        "user_id",
        "first_name",
        "last_name",
        "gender",
        "email",
        "phone",
        "date_of_birth",
        "job_title"
    ]
]
```

### STEP 4: MISSING VALUE HANDLING

```python
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())
```

```python
# Numerical columns
numeric_cols = df.select_dtypes(include=["number"]).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Categorical columns
categorical_cols = df.select_dtypes(include=["object"]).columns

for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

### STEP 5: DATA TYPE CORRECTION

```python
df["date_of_birth"] = pd.to_datetime(
    df["date_of_birth"],
    errors="coerce"
)
```

### STEP 6: FORMAT STANDARDIZATION

```python
df["first_name"] = df["first_name"].str.strip().str.lower()
df["last_name"] = df["last_name"].str.strip().str.lower()
df["email"] = df["email"].str.strip().str.lower()
df["job_title"] = df["job_title"].str.strip().str.lower()

# Standardize gender values
df["gender"] = df["gender"].replace({
    "M": "Male",
    "F": "Female",
    "male": "Male",
    "female": "Female"
})
```

```python
# Remove spaces from phone numbers
df["phone"] = df["phone"].astype(str).str.strip()
```

### STEP 7: VALIDATION


```python
assert df.duplicated().sum() == 0
print("\nDuplicates Remaining:", df.duplicated().sum())
```
